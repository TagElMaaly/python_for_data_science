# 2 Filter dataframe based on a multiple conditions

```python
# find Apple laptops that cost more than 70000 euros
(df_laptops['Company']=='Apple') & (df_laptops['Price']>70000)
```

## 2.1 & (and)

```python
# filter datframe based on multiple conditions

df_laptops[(df_laptops['Company']=='Apple') & (df_laptops['Price']>70000)]
```

```python
and -> &
or -> |
```

## 2.2 | (or)

```python
# find laptops Apple or Dell laptops
df_laptops['Company']=='Apple'
df_laptops['Company']=='Dell'

(df_laptops['Company']=='Apple') | (df_laptops['Company']=='Dell')
```

```python
# filter dataframe based on multiple conditions
df_laptops[(df_laptops['Company']=='Apple') | (df_laptops['Company']=='Dell')].value_counts('Company')
```

```python
# find laptops from Apple or Dell that cost over 70000 euros
df_laptops['Company']=='Apple'
df_laptops['Company']=='Dell'
df_laptops['Price']>70000

((df_laptops['Company']=='Apple') | (df_laptops['Company']=='Dell')) & (df_laptops['Price']>70000)
```

```python
# filter dataframe based on multiple conditions

df_laptops[((df_laptops['Company']=='Apple') | (df_laptops['Company']=='Dell')) & (df_laptops['Price']>70000)]
```
