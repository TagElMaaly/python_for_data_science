# 1 isin(): Single Filtering

```python
import pandas as pd
```

```python
df_laptops = pd.read_csv('laptop_price.csv')
```

```python
df_laptops.head(3)
```

```python
# select Applw or HP Laptops
df_laptops['Company'].isin(['Apple', 'HP'])
```

```python
# filter dataframes
df_laptops[df_laptops['Company'].isin(['Apple', 'HP'])].value_counts('Company')
```

# 2 isin(): Multiple Filtering

```python
# find notebooks or ultabooks from Apple or HP
filter_1 = df_laptops['TypeName'].isin(['Notebook', 'Ultrabook'])
filter_2 = df_laptops['Company'].isin(['Apple', 'HP'])
```

```python
# filter dataframe
df_laptops[filter_1 & filter_2]
```

```python
df_laptops[filter_1 | filter_2]
```
