# 1 Creating a conditional Column from More than 2 choices: np.select()

```python
import numpy as np
```

```python
import numpy as np
```

```python
# Create an array based on multible price tiers (+2 choices)

# create "conditoins" and "values" variables
conditions = [
    df_laptops['Price']>3000,
    (df_laptops['Price']>2000) & (df_laptops['Price']<=3000),
    (df_laptops['Price']>800) & (df_laptops['Price']<=2000),
    df_laptops['Price']<=800
]

values = ['Too Expensive', 'Expensive', 'Affodable', 'Cheap']
```

```python
# Set it to new column
df_laptops['Price_tiers'] = np.select(conditions, values)
```

```python
col         = 'Price'
conditions  = [ df_laptops[col] >= 400, (df_laptops[col] < 400) & (df_laptops[col]> 200), df_laptops[col] <= 200 ]
choices     = [ "high", 'medium', 'low' ]
    
df_laptops["energy_class"] = np.select(conditions, choices, default=np.nan)
```

```python
df_laptops['Price_tires'] = np.where(
    df_laptops['Price'] > 70000,  "Too Expensive",
    np.where(
        (df_laptops['Price'] <= 70000) & (df_laptops['Price'] > 50000),  "Expensive",
             np.where(
                 (df_laptops['Price'] <= 50000) & (df_laptops['Price'] > 30000),  "Affodable",
                     "Cheap"
             )
    )
)
df_laptops
```

```python
# count values inside "price_tier" column
df_laptops.value_counts('Price_tires')
```
