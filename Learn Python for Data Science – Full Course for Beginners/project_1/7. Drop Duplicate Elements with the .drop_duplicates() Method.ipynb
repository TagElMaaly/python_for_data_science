# 1 drop_duplicates()

```python
# drop duplicates in 2 or more columns
df_laptops.drop_duplicates(['Company'])[['Company', 'Price']]
```

```python
# sort dataframe asending by company and price
# cheapest first and most expensive last
df_laptops = df_laptops.sort_values(['Company', 'Price'])
```

```python
# cheapest: keep='last'
df_laptops.drop_duplicates(['Company'], keep='first')[['Company', 'Price']]
```

```python
# most expensive: keep='last'
df_laptops.drop_duplicates(['Company'], keep='last', inplace=True, ignore_index=True)
```

```python
# Arguments

# inplace: drop duplicates in place or to return a copy.
# ignore_index=True: the resulting axis will be labeled 0, 1, ...
df_laptops[['Company', 'Price']]
```
