# 1 Create a column based on 1 conditions:
# np.where()

```python
import numpy as np
```

```python
# create an arry bassed on proce tiers
np.where(df_laptops['Price'] > 70000,  "Expensive", "Cheap")
```

```python
# set it to a new column
df_laptops["Price_tier"] = np.where(df_laptops['Price'] > 70000,  "Expensive", "Cheap")
```

```python
# show first 5 rows
```

```python
# count values in price_tier coulmn
df_laptops.value_counts('Price_tier')
```
