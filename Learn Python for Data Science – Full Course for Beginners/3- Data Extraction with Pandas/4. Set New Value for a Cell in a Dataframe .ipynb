```python
import pandas as pd
```

```python
# reads csv file
df =  pd.read_csv('players_20.csv')
#set index
df.set_index('short_name', inplace=True)
# selecr column
df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

```python
df
```

# 1 Set value to one cell


```python
# update L. Messi height
# df.iloc[[0], [3]] = 180
df.loc['L. Messi', 'height_cm'] = 175
```

```python
df
```

# 2 Set value to entire column

```python
# set height of all players to 190
# df[['height_cm']] = 200
df.loc[:, 'height_cm'] = 190
```

```python
df
```

# 3 Set value to entire row

```python
# get all columns that correspond to player ranked last in fifa
# df.iloc[18277, :]
# df.iloc[[-1], :] = ""
df.iloc[[-1], :]
```

```python
# set null values to all columns that corresponds to player ranked last
import numpy as np

df.iloc[[-1], :] = np.nan
```

```python
df
```

# 4 Set values to Multiple cells

```python
# set values for items matching the list of labels
df.loc[['L. Messi', 'Cristiano Ronaldo'], ['height_cm']] = 175
```

```python
df
```

# 5 Set value for rows matching a condition

```python
# set value for rows matching a condition
columns = ['age', 'dob', 'height_cm', 'weight_kg']
# df[df['age']==22][['club']] = "Test"  # it through error: A value is trying to be set on a copy of a slice from a DataFrame.
# Try using .loc[row_indexer,col_indexer] = value instead
# df.loc[df['age']==22, ['club']] = "Test"
# df.loc[df['age']==22, ['club']]
df.loc[df['height_cm'] > 180, columns] = 0
```
