```python
import pandas as pd
```

```python
# reads csv file
df =  pd.read_csv('players_20.csv')
#set index
df.set_index('short_name', inplace=True)
# selecr column
df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

```python
df
```

# 1 copy()


#### 1.1 deep = True

```python
# deep=True by default (Modifications to the data or indices of the copy
df_copy = df.copy()
```

```python
# update value on original dataframe
df.loc['L. Messi', 'height_cm'] = 180
```

```python
df
```

```python
df_copy
```

```python
# deep=False (Any changes to the data of the original will be reflected in the shallow copy and vice verrsa)
df_shallow_copy = df.copy(deep=False)
```

```python
df.loc['Cristiano Ronaldo', 'height_cm'] = 200
```

```python
# shallow copy vs original dataframe
df
```

```python
df_shallow_copy
```

#### 1.3 Simple assignment

```python
# make a copy
df_new_copy = df.copy()
```

```python
# update value in original dataframe
df.loc['Neymar Jr', 'height_cm'] = 300
```

```python
# copy vs original dataframe
df
```
