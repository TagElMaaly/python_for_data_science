# 2 Read .csv from Multiple URLs with Pandas

- https://www.football-data.co.uk/mmz4281/2526/E0.csv
- https://www.football-data.co.uk/mmz4281/2526/E1.csv

#### Link: root + season + league

```python
# link structure
"https://www.football-data.co.uk/mmz4281/" + "2526" + "/" + "E0" + ".csv"
```

```python
# create a root variable
root = "https://www.football-data.co.uk/mmz4281/"
```

## 2.1 Multiple leagues

- https://www.football-data.co.uk/mmz4281/2526/E0.csv
- https://www.football-data.co.uk/mmz4281/2526/E1.csv
- https://www.football-data.co.uk/mmz4281/2526/E2.csv
- https://www.football-data.co.uk/mmz4281/2526/E3.csv
- https://www.football-data.co.uk/mmz4281/2526/EC.csv

```python
# creat list of leagues
leagues = ['E0', 'E1', 'E2', 'E3', 'EC']
frames =[]

# lopping through leagues, read multiple csv and append it into a list
for league in leagues:
    df = pd.read_csv(root + '2526' + '/' + league + '.csv', encoding='unicode_escape')
    frames.append(df)
```

```python
# length of frames
len(frames)
```

```python
# show 1st, 2nd and 3rd element
frames[4]
```

## 2.2 Multipe Seasons

```python
for season in range(15, 26):
    print(str(season) + str(season + 1))
```

```python
# creat list of leagues
leagues = ['E0', 'E2', 'E3']
frames =[]

# loop through range of seasons and insert new column
for league in leagues:
    for season in range(15, 21):
        df = pd.read_csv(root + '2526' + '/' + league + '.csv')
        df.insert(1, 'season', season)
        frames.append(df)

        
```

```python
# total frames (6 seasons x 3 leagues)
len(frames)
```

```python
# show 1st, and last element (new season column was created)
frames[-1]
```

# 3 Organize All The Data in a Dictionary

```python
# create a dictionary with original name of league as key
dict_countries = {
    'Spanish La Liga': 'SP1',
    'Spanish Segunda Division': 'SP2',
    'German Bundes liga': 'D1',
    'English Primier League': 'E0',
    'English League 1': 'E2',
    'English League 2': 'E3',
}
```

```python
# get a dictionary element
dict_countries['Spanish La Liga']
```

```python
for league in dict_countries:
    print(dict_countries[league])
```

```python
dict_historical_data = {}

# creating list through key elements
# leagues = ['E0', 'E2', 'E3']

# looping through key elements
for league in dict_countries:
    frames =[]
    for season in range(15, 21):
        df = pd.read_csv(root + str(season) + str(season+1) + '/' + dict_countries[league] + '.csv')
        df.insert(1, 'season', season)
        frames.append(df)
    df_concat = pd.concat(frames)
    dict_historical_data[league] = df_concat
```
