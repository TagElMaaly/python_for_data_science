```python
import pandas as pd
import numpy as np
```

```python
df_laptops = pd.read_csv('laptop_price.csv')
```

```python
df_laptops.keys()
```

```python
df_laptops = df_laptops.rename(columns={'Price: 0': "Price_euros"})
```

```python
df_laptops
```

# 1 Filter dataframe based on a condition

```python
# find which row have "Apple" in the "Company" column
df_laptops['Company']=='Apple'
```

```python
# filter dataframe based on condition
df_laptops[df_laptops['Company']=='Apple']
```

```python
# verify which companies are in the "Company" column with value_counts()
df_laptops[df_laptops['Company']=='Apple'].value_counts('Company')
```
